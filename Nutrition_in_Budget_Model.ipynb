```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pickle

# Load the dataset
df = pd.read_excel('/content/recipeeees.xlsx')

```

```python
df
```

```python
# Features for clustering
X = df[['calories', 'protein', 'carbs', 'fat', 'price($)']]

```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
X_scaled
```

```python
# Train a K-Means model
n_clusters = 10  # Number of clusters can be adjusted
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X_scaled)

```

```python
# Save the model and scaler to .sav files
with open('kmeans_model.sav', 'wb') as model_file:
    pickle.dump(kmeans, model_file)
with open('scaler_model.sav', 'wb') as scaler_file:
    pickle.dump(scaler, scaler_file)

print("Model and scaler saved successfully.")
```
